```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

**1.BOX PLOT**

```python
sns.boxplot(data = penguins, x = 'species', y ='body_mass_g', hue = "sex")
```

**2.VILION PLOT**

```python
sns.violinplot(data = penguins, x = 'species', y ='body_mass_g', hue = "sex")
```

```python
# Width of the violin: Density of the data at different values
# Middle line: Median
# Thick bar: Interquartile range (if inner='box')
```

**KDE PLOT**

```python
# KDE -> for estimating the probability density function of continuous or non-parametric data
# Or in simple terms: a smooth curve that shows the distribution of data
# It helps you see where the data is concentrated (peaks) and how it spreads.

sns.kdeplot(data = penguins, x = "body_mass_g", hue = "species", palette = "pastel", fill = "True") 
```

**Heatmap**

```python
# Heatmap: a 2D color-coded matrix that shows relationships or patterns between
# two variables.
# It's commonly used to:
# - Show correlation between features
# - Visualize confusion matrices -> will see in the later stages.
# - Highlight missing data -> sns.heatmap(df.isnull())
```

```python
colums = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g" ]
penguins[colums].corr()
```

```python
sns.heatmap(data = penguins[colums].corr(), annot = True)
```

**RUG PLOT**

```python
# Rug plot:
sns.rugplot(data = penguins, y = "body_mass_g", hue = "species", palette = "Set2", height = 0.05)
```

**Pair Plot**

```python
sns.pairplot(data = penguins,hue ="sex", diag_kind ="kde", )
```

**Pair Grid**

```python
#pair grid -> subplot grid for plotting pairwise relationships in a dataset.
graph = sns.PairGrid(data = penguins, hue = "sex", palette = "Set2")
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
```
