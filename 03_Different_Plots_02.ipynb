```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

**1.Histogaram**

```python
#seting contexr
sns.set_context("notebook")

#Ploting the graph
sns.histplot(data=penguins, x="body_mass_g", hue = "sex", palette="Set3")
```

```python
# Using mutilple to ci=ombine the garph
sns.histplot(data=penguins, x="body_mass_g", hue = "sex", palette="Set3", multiple="stack")
```

**2.Regression Plot**


```python
sns.regplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm")
```

**3.Line Plot**

```python
sns.set_style("whitegrid")
sns.lineplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", hue = "sex", style = "island")
```

**4.JOIN PLOT**

```python
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm", hue="island")
```

**5.Bar Plot**

```python
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette=["yellow", "red"])
```

```python
# using estimator:
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette=["yellow", "red"], estimator=np.sum)
```

**6.Count Plot**

```python
sns.countplot(data = penguins, x = "species")
```
