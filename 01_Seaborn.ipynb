## SEABORN
Seaborn is a data visualization library built on top of Matplotlib. It is often used because it makes attractive visualizations and works easily with Pandas. While in Matplotlib you often had to write multiple lines of code to create a plot Seaborn makes assumptions on what you want which often translates into getting the same plot with 1 line of code 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Import data**

```python
# You can import custom data
cs_df = pd.read_csv('ComputerSales.csv')

# Seaborn provides built in datasets
print(sns.get_dataset_names())

# Load a built in dataset based on US State car crash percentages
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# Count of different species and island
print(penguins["species"].value_counts())
print(penguins["island"].value_counts())
```

**The way of styling themes are as follows:** 

**Syntax : sns.set_style("white")**

1. white
2. dark
3. whitegrid
4. darkgrid
5. ticks

**set_context() : allows us to override default parametres. This affect the things like size of labels, lines and other elements of the plot, but not overall style. The context are:**

1. paper -> small elemnt , ideal for printing in academic papers.
2. notebook -> medium size, good for working inside jupyter
3. talk -> Large text and lines, good for slides
4. poster -> Very large elements, suitable for large display



**palette : used for seting colors**
1. Set 1 : bright and high contrast colors : 9
2. Set 2 : Softer, more pastel like colors : 8
3. Set 3 : A mix of soft and bold, varied hues : 12

**Also have paraeters**
1. deep : Default palette, rich and balanced
2. muted : Softer colors, good for categories
3. bright : High contrast, bold colors
4. pastel : Soft, pale tones
5. dark : Darker tones
6. colorblind : optimized for colorbling

**Palette Type:**
1. Qualitative : Categorial in nature, colour asssigned to each group must be distinct.
2. Sequential : Data with ordered progression (low -> high)
3. Diveging : Data with a meaningful midpoint or mixed values
