```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**loading Data**

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

**1.Scatter Plot**

```python
# Seting the style
sns.set_style("dark")

#Ploting a graph
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g" , hue = "island", palette = "Dark2")

#used to remove spine
sns.despine()

#set the context
sns.set_context("notebook")
```

```python
#using style : Different categories ko easily identify karne ke liye
sns.scatterplot(data = penguins, x = "species", y = "body_mass_g" , hue = "island", style = "sex")
```

```python
#it is ovelaping a lot 
#Using alpha to decrease the opacity
sns.scatterplot(data = penguins, x = "species", y = "body_mass_g" , hue = "island", style = "sex" , alpha = 0.5)
```

**2.STRIP PLOT** : Better representation of data spread

```python
sns.stripplot(data = penguins, x = "species", y = "body_mass_g" , hue = "island")
```

```python
# Using Dodge equal to true to seprate data points
sns.stripplot(data = penguins, x = "species", y = "body_mass_g" , hue = "island", dodge=True)
```

```python
##Using jitter for better readabilitys
sns.stripplot(data = penguins, x = "species", y = "body_mass_g" , hue = "island" , dodge = True, jitter = True)
```

**3.SWARM PLOT: Print data in shape of a swarm** 

```python
sns.swarmplot(data = penguins, x = "species", y = "body_mass_g" , hue = "island")
```
